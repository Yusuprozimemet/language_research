```python
# =====================================================
#  GENELINGUA v3.5 ‚Äì Evidence-Based from Published GWAS
#  -------------------------------------------------
#  ‚Ä¢ SNPs and betas from actual published studies
#  ‚Ä¢ Realistic effect sizes from peer-reviewed research
#  ‚Ä¢ Proper PGS construction methodology
#  ‚Ä¢ Clear evidence grading and ancestry notes
#  ‚Ä¢ Comprehensive citations
#  ‚≠ê UPDATED: HARDCODED ANCESTRY SELECTION ‚≠ê
#  -------------------------------------------------
#  ‚Ä¢ All SNPs now have beta_EUR, beta_EAS, beta_default
#  ‚Ä¢ rs4680 (COMT) marked as ancestry_specific with OPPOSITE effects
#  ‚Ä¢ compute_pgs() properly uses selected ancestry betas
#  ‚Ä¢ No widgets - simple hardcoded ancestry selection
#  ‚Ä¢ Change SELECTED_ANCESTRY = 'EUR' to modify ancestry
#  -------------------------------------------------
# =====================================================

# ‚îÄ‚îÄ 1. Install & import ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import math
import json
import re
import zipfile
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import io
from matplotlib.patches import Patch
from IPython.display import display, Markdown, HTML
from google.colab import files
from scipy.stats import norm
from datetime import datetime, timezone
!pip install - q pandas numpy matplotlib seaborn scipy


sns.set_style("whitegrid")
%matplotlib inline

# ‚îÄ‚îÄ 2. SNP Database - Evidence-Based ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Based on published GWAS studies with actual effect sizes
SNP_DB_PGS = {
    # COMT Gene - Dopamine metabolism (ANCESTRY-SPECIFIC!)
    'rs4680': {
        'gene': 'COMT',
        'label': 'COMT Val158Met (rs4680)',
        'effect_allele': 'A',  # Met allele
        'beta_EUR': 0.042,   # Met allele beneficial in Europeans
        'beta_EAS': -0.042,  # Val allele beneficial in East Asians (OPPOSITE!)
        'beta_default': 0.025,  # Fallback for other ancestries
        'ancestry_specific': True,  # Flag for ancestry-dependent effects
        'se': 0.015,
        'p_value': 0.006,
        'evidence': 'Strong',
        'phenotype': 'Second language learning (white matter & phonetic learning)',
        'population': 'European ancestry adults',
        'refs': ['Mamiya2016', 'Colzato2010', 'Bonetti2021'],
        'notes': 'Met allele beneficial in EUR, Val allele beneficial in EAS - major ancestry difference!'
    },
    'rs737865': {
        'gene': 'COMT',
        'label': 'COMT rs737865',
        'effect_allele': 'C',
        'beta_EUR': 0.018,
        'beta_EAS': 0.018,
        'beta_default': 0.018,
        'ancestry_specific': False,
        'se': 0.008,
        'p_value': 0.024,
        'evidence': 'Moderate',
        'phenotype': 'Figural fluency/creativity',
        'population': 'Chinese college students',
        'refs': ['Zhang2014'],
        'notes': 'Associated with creative thinking, which relates to language flexibility'
    },

    # Reading & Language Processing
    'rs76551189': {
        'gene': 'Near IGSF9',
        'label': 'Nonword reading (rs76551189)',
        'effect_allele': 'A',
        'beta_EUR': 0.025,
        'beta_EAS': 0.025,
        'beta_default': 0.025,
        'ancestry_specific': False,
        'se': 0.005,
        'p_value': 5.2e-8,
        'evidence': 'Strong',
        'phenotype': 'Nonword reading ability',
        'population': 'Meta-analysis (34,000 individuals)',
        'refs': ['Eising2022'],
        'notes': 'Genome-wide significant for phonological processing'
    },
    'rs4656859': {
        'gene': 'IGSF9',
        'label': 'IGSF9 rs4656859',
        'effect_allele': 'T',
        'beta_EUR': 0.022,
        'beta_EAS': 0.022,
        'beta_default': 0.022,
        'ancestry_specific': False,
        'se': 0.005,
        'p_value': 8.1e-8,
        'evidence': 'Strong',
        'phenotype': 'Nonword reading',
        'population': 'Meta-analysis',
        'refs': ['Eising2022'],
        'notes': 'Phonological decoding - key for language learning'
    },
    'rs1422268': {
        'gene': 'DOK3',
        'label': 'DOK3 rs1422268',
        'effect_allele': 'A',
        'beta_EUR': 0.020,
        'beta_EAS': 0.020,
        'beta_default': 0.020,
        'ancestry_specific': False,
        'se': 0.005,
        'p_value': 1.3e-7,
        'evidence': 'Strong',
        'phenotype': 'Nonword reading',
        'population': 'Meta-analysis',
        'refs': ['Eising2022'],
        'notes': 'Reading-related cognitive processing'
    },

    # Phonemic Awareness
    'rs765166': {
        'gene': 'ZFAS1',
        'label': 'ZFAS1 rs765166',
        'effect_allele': 'C',
        'beta_EUR': 0.015,
        'beta_EAS': 0.015,
        'beta_default': 0.015,
        'ancestry_specific': False,
        'se': 0.006,
        'p_value': 0.012,
        'evidence': 'Moderate',
        'phenotype': 'Phonemic awareness',
        'population': 'Meta-analysis',
        'refs': ['Eising2022'],
        'notes': 'Phoneme manipulation - critical for pronunciation'
    },

    # Memory & Cognitive Processing
    'rs363039': {
        'gene': 'SNAP25',
        'label': 'SNAP-25 rs363039',
        'effect_allele': 'G',
        'beta_EUR': 0.028,
        'beta_EAS': 0.028,
        'beta_default': 0.028,
        'ancestry_specific': False,
        'se': 0.011,
        'p_value': 0.011,
        'evidence': 'Moderate',
        'phenotype': 'Performance IQ (visuospatial)',
        'population': 'Dutch children & adults',
        'refs': ['Gosso2006'],
        'notes': 'Synaptic protein - learning & memory functions'
    },

    # Working Memory
    'rs80263879': {
        'gene': 'EPHX2',
        'label': 'EPHX2 rs80263879',
        'effect_allele': 'G',
        'beta_EUR': 0.035,
        'beta_EAS': 0.035,
        'beta_default': 0.035,
        'ancestry_specific': False,
        'se': 0.008,
        'p_value': 2.1e-8,
        'evidence': 'Strong',
        'phenotype': 'Static spatial working memory',
        'population': 'Chinese students',
        'refs': ['Zhang2022'],
        'notes': 'Genome-wide significant for working memory'
    },

    # General Cognitive Ability (g-factor)
    'rs11264236': {
        'gene': 'NPR1',
        'label': 'NPR1 rs11264236',
        'effect_allele': 'A',
        'beta_EUR': 0.012,
        'beta_EAS': 0.012,
        'beta_default': 0.012,
        'ancestry_specific': False,
        'se': 0.005,
        'p_value': 0.016,
        'evidence': 'Moderate',
        'phenotype': 'Science academic attainment',
        'population': 'UK adolescents (ALSPAC)',
        'refs': ['Donati2021'],
        'notes': 'Academic performance - language component'
    },
    'rs10905791': {
        'gene': 'ASB13',
        'label': 'ASB13 rs10905791',
        'effect_allele': 'T',
        'beta_EUR': 0.011,
        'beta_EAS': 0.011,
        'beta_default': 0.011,
        'ancestry_specific': False,
        'se': 0.005,
        'p_value': 0.028,
        'evidence': 'Moderate',
        'phenotype': 'Science attainment',
        'population': 'UK adolescents',
        'refs': ['Donati2021'],
        'notes': 'Cognitive performance in structured learning'
    },

    # Auditory Processing
    'rs6453022': {
        'gene': 'ARHGEF28',
        'label': 'ARHGEF28 rs6453022',
        'effect_allele': 'C',
        'beta_EUR': 0.016,
        'beta_EAS': 0.016,
        'beta_default': 0.016,
        'ancestry_specific': False,
        'se': 0.004,
        'p_value': 3.2e-9,
        'evidence': 'Strong',
        'phenotype': 'Hearing difficulty (inverse)',
        'population': 'UK Biobank (250,389 individuals)',
        'refs': ['Wells2019'],
        'notes': 'Auditory processing - crucial for language perception'
    },
    'rs9493627': {
        'gene': 'EYA4',
        'label': 'EYA4 rs9493627',
        'effect_allele': 'G',
        'beta_EUR': 0.014,
        'beta_EAS': 0.014,
        'beta_default': 0.014,
        'ancestry_specific': False,
        'se': 0.004,
        'p_value': 1.8e-8,
        'evidence': 'Strong',
        'phenotype': 'Hearing function',
        'population': 'UK Biobank',
        'refs': ['Wells2019'],
        'notes': 'Auditory system - language input processing'
    },

    # Hippocampal Volume (Memory structure)
    'rs7294919': {
        'gene': 'ASTN2',
        'label': 'ASTN2 rs7294919',
        'effect_allele': 'T',
        'beta_EUR': 0.019,
        'beta_EAS': 0.019,
        'beta_default': 0.019,
        'ancestry_specific': False,
        'se': 0.004,
        'p_value': 2.3e-8,
        'evidence': 'Strong',
        'phenotype': 'Hippocampal volume',
        'population': 'Meta-analysis (33,536 individuals)',
        'refs': ['Hibar2017'],
        'notes': 'Memory formation - essential for vocabulary acquisition'
    },

    # DRD2 - Reward & motivation
    'rs1800497': {
        'gene': 'DRD2/ANKK1',
        'label': 'DRD2 TaqIA (rs1800497)',
        'effect_allele': 'T',  # A1 allele
        'beta_EUR': 0.015,
        'beta_EAS': 0.015,
        'beta_default': 0.015,
        'ancestry_specific': False,
        'se': 0.007,
        'p_value': 0.032,
        'evidence': 'Moderate',
        'phenotype': 'Working memory span',
        'population': 'Chinese adults',
        'refs': ['Gong2012'],
        'notes': 'Dopamine receptor - motivation & reward in learning'
    },

    # Brain Aging (Cognitive maintenance)
    'rs10009513': {
        'gene': 'MAPT-AS1',
        'label': 'MAPT-AS1 rs10009513',
        'effect_allele': 'A',
        # From Ning 2021: brain aging (negative = younger brain)
        'beta_EUR': -0.022,
        'beta_EAS': -0.022,
        'beta_default': -0.022,
        'ancestry_specific': False,
        'se': 0.005,
        'p_value': 1.4e-8,
        'evidence': 'Strong',
        'phenotype': 'Relative brain age (protective)',
        'population': 'UK Biobank',
        'refs': ['Ning2021'],
        'notes': 'Neuroprotective - maintains cognitive function over time'
    }
}

# ‚îÄ‚îÄ 3. Comprehensive References ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
REFERENCES = {
    "Mamiya2016": {
        "citation": "Mamiya PC, Richards TL, Coe BP, Eichler EE, Kuhl PK (2016). Brain white matter structure and COMT gene are linked to second-language learning in adults. PNAS 113(26):7249-7254.",
        "url": "https://www.pnas.org/doi/full/10.1073/pnas.1606602113",
        "pmid": "27298360",
        "key_finding": "COMT Met allele associated with better phonetic learning and white matter integrity in L2 learners"
    },
    "Eising2022": {
        "citation": "Eising E, et al (2022). Genome-wide analyses of individual differences in quantitatively assessed reading- and language-related skills in up to 34,000 people. PNAS 119(35):e2202764119.",
        "url": "https://www.pnas.org/doi/10.1073/pnas.2202764119",
        "pmid": "35998220",
        "key_finding": "Large-scale GWAS identified 42 genome-wide significant loci for reading/language traits. GenLang Consortium study."
    },
    "Wells2019": {
        "citation": "Wells HRR, et al (2019). GWAS Identifies 44 Independent Associated Genomic Loci for Self-Reported Adult Hearing Difficulty in UK Biobank. Am J Hum Genet 105(4):788-802.",
        "url": "https://www.cell.com/ajhg/fulltext/S0002-9297(19)30334-0",
        "pmid": "31564434",
        "key_finding": "44 loci associated with hearing - critical for language input processing"
    },
    "Hibar2017": {
        "citation": "Hibar DP, et al (2017). Novel genetic loci associated with hippocampal volume. Nat Commun 8:13624.",
        "url": "https://www.nature.com/articles/ncomms13624",
        "pmid": "28098162",
        "key_finding": "6 loci for hippocampal volume - memory structure essential for vocabulary"
    },
    "Zhang2022": {
        "citation": "Zhang L, et al (2022). A genome-wide association study identified one variant associated with static spatial working memory in Chinese population. Brain Sci 12(9):1254.",
        "url": "https://www.mdpi.com/2076-3425/12/9/1254",
        "pmid": "36176292",
        "key_finding": "EPHX2 variant genome-wide significant for working memory"
    },
    "Gosso2006": {
        "citation": "Gosso MF, et al (2006). The SNAP-25 gene is associated with cognitive ability: evidence from a family-based study in two independent Dutch cohorts. Mol Psychiatry 11:878-886.",
        "url": "https://www.nature.com/articles/4001868",
        "pmid": "16801949",
        "key_finding": "SNAP-25 variants associated with performance IQ"
    },
    "Zhang2014": {
        "citation": "Zhang S, et al (2014). Association of COMT and COMT-DRD2 interaction with creative potential. Front Hum Neurosci 8:216.",
        "url": "https://www.frontiersin.org/articles/10.3389/fnhum.2014.00216",
        "pmid": "24782743",
        "key_finding": "COMT variants linked to creative fluency - relevant to flexible language use"
    },
    "Gong2012": {
        "citation": "Gong P, et al (2012). An association study on the polymorphisms of dopaminergic genes with working memory in a healthy Chinese Han population. J Neural Transm 119:1293-1302.",
        "url": "https://link.springer.com/article/10.1007/s00702-012-0817-9",
        "pmid": "22362150",
        "key_finding": "DRD2 TaqIA associated with working memory span"
    },
    "Donati2021": {
        "citation": "Donati G, et al (2021). Evidence for specificity of polygenic contributions to attainment in English, maths and science during adolescence. Learn Individ Differ 88:101993.",
        "url": "https://www.sciencedirect.com/science/article/pii/S1041608021000674",
        "pmid": "33594131",
        "key_finding": "Subject-specific genetic contributions to academic attainment"
    },
    "Ning2021": {
        "citation": "Ning K, et al (2021). Improving brain age estimates with deep learning leads to identification of novel genetic factors associated with brain aging. Neurobiol Aging 102:89-97.",
        "url": "https://www.sciencedirect.com/science/article/pii/S0197458021001123",
        "pmid": "34098431",
        "key_finding": "80 SNPs associated with brain aging - cognitive maintenance"
    },
    "Colzato2010": {
        "citation": "Colzato LS, et al (2010). The flexible mind is associated with the catechol-O-methyltransferase (COMT) Val158Met polymorphism. Neuropsychologia 48(1):220-225.",
        "pmid": "20434465",
        "key_finding": "COMT affects cognitive flexibility"
    },
    "Bonetti2021": {
        "citation": "Bonetti L, et al (2021). Brain predictive coding processes are associated to COMT gene Val158Met polymorphism. NeuroImage 233:117954.",
        "pmid": "33716157",
        "key_finding": "COMT heterozygotes show stronger neural response to auditory deviants"
    }
}

# ‚îÄ‚îÄ 4. Critical Disclaimers ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(HTML("""
<div style="background-color: #fff3cd; border-left: 5px solid #ffc107; padding: 15px; margin: 20px 0;">
<h3 style="color: #856404; margin-top: 0;">‚ö†Ô∏è SCIENTIFIC LIMITATIONS</h3>
<ul style="color: #856404;">
<li><b>No validated \"language learning PGS\" exists.</b> These SNPs come from studies of reading, memory, hearing, and cognitive ability - not language learning per se.</li>
<li><b>Small effect sizes:</b> Combined, these variants explain <b>~2-4% of variance</b> in related cognitive traits.</li>
<li><b>Ancestry matters critically:</b> Most studies are European-ancestry. COMT shows <b>opposite effects</b> in East Asian populations.</li>
<li><b>Environment >> Genetics:</b> Study method, motivation, immersion time, and instruction quality are 20-50x more important.</li>
<li><b>Cross-trait inference:</b> We're inferring language learning from reading/memory/hearing genetics.</li>
<li><b>Educational tool only:</b> Not diagnostic, not predictive of individual success.</li>
</ul>
</div>
"""))

# ‚îÄ‚îÄ 5. Ancestry Configuration ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# ============================================================
# üîß MODIFY THIS SECTION TO CHANGE ANCESTRY
# ============================================================
# Available options: 'EUR', 'EAS', 'SAS', 'AFR', 'AMR', 'MENA', 'OTH'
# EUR = European (includes European-American, European-descended populations)
# EAS = East Asian (Chinese, Japanese, Korean, Vietnamese, etc.)
# SAS = South Asian (Indian, Pakistani, Bangladeshi, Sri Lankan, etc.)
# AFR = African (Sub-Saharan African, African-American)
# AMR = Latino/Admixed American (Mexican, Puerto Rican, Brazilian, etc.)
# MENA = Middle Eastern/North African
# OTH = Other/Mixed (results will be highly uncertain)

SELECTED_ANCESTRY = 'EAS'  # üëà CHANGE THIS LINE TO MODIFY ANCESTRY

# ============================================================

# Display ancestry selection
ancestry_labels = {
    'EUR': 'European (includes European-American, European-descended populations)',
    'EAS': 'East Asian (Chinese, Japanese, Korean, Vietnamese, etc.)',
    'SAS': 'South Asian (Indian, Pakistani, Bangladeshi, Sri Lankan, etc.)',
    'AFR': 'African (Sub-Saharan African, African-American)',
    'AMR': 'Latino/Admixed American (Mexican, Puerto Rican, Brazilian, etc.)',
    'MENA': 'Middle Eastern/North African',
    'OTH': 'Other/Mixed (results will be highly uncertain)'
}

selected_label = ancestry_labels.get(SELECTED_ANCESTRY, SELECTED_ANCESTRY)

display(HTML(f"""
<div style="background-color: #d4edda; border-left: 5px solid #28a745; padding: 20px; margin: 20px 0;">
<h2 style="color: #155724; margin-top: 0;">üß¨ Ancestry Configuration</h2>
<p style="color: #155724; font-size: 14pt;">
<b>Selected Ancestry:</b> {selected_label}<br><br>
<b>‚ö° Key Impact for rs4680 (COMT):</b><br>
{"‚Ä¢ EUR: Met allele beneficial (+0.042)" if SELECTED_ANCESTRY == 'EUR' else "‚Ä¢ EAS: Val allele beneficial (-0.042)" if SELECTED_ANCESTRY == 'EAS' else "‚Ä¢ Using default approximation"}<br><br>
<b>To change ancestry:</b> Modify SELECTED_ANCESTRY = '{SELECTED_ANCESTRY}' in the code above
</p>
</div>
"""))

if SELECTED_ANCESTRY in ['AFR', 'SAS', 'AMR', 'MENA', 'OTH']:
    display(HTML(f"""
    <div style="background-color: #fff3cd; border-left: 5px solid #ff9800; padding: 15px; margin: 10px 0;">
    <p style="color: #e65100; font-weight: bold; font-size: 14pt;">
    ‚ö†Ô∏è WARNING: You selected "{selected_label}"
    </p>
    <p style="color: #e65100;">
    <b>Almost all SNPs in this analysis were discovered in European samples.</b><br>
    For your ancestry:<br>
    ‚Ä¢ Effect sizes may be completely different<br>
    ‚Ä¢ Some SNPs may not be polymorphic in your population<br>
    ‚Ä¢ Linkage disequilibrium patterns differ<br>
    ‚Ä¢ <b>Results should be treated as HIGHLY SPECULATIVE</b><br><br>
    The analysis will continue using European betas as a rough approximation,
    but <b>do not trust the percentile estimates</b> for your ancestry group.
    </p>
    </div>
    """))

print(f"‚úì Ancestry set to: {SELECTED_ANCESTRY} ({selected_label})")
print("=" * 60)

# ‚îÄ‚îÄ 6. Upload & Parse 23andMe ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ


def load_23andme():
    uploaded = files.upload()
    name, data = next(iter(uploaded.items()))
    if name.lower().endswith('.zip'):
        with zipfile.ZipFile(io.BytesIO(data)) as z:
            txt_files = [f for f in z.namelist() if f.lower().endswith('.txt')]
            if not txt_files:
                raise ValueError("No .txt file found in ZIP.")
            txt = txt_files[0]
            raw = z.read(txt).decode('utf-8', errors='ignore')
    else:
        raw = data.decode('utf-8', errors='ignore')
    return raw, name


print("\nüìÅ STEP 2: Upload your 23andMe raw data file (.txt or .zip):")
print("   (This upload will appear AFTER you confirm ancestry above)")
raw_text, fname = load_23andme()
print(f"‚úì Loaded: {fname}")

# Ancestry is now hardcoded above - no need for container logic
print(f"üîç DEBUG: Using SELECTED_ANCESTRY = '{SELECTED_ANCESTRY}'")

# ‚îÄ‚îÄ 6. Parse SNPs ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ


def parse_23andme(txt):
    snp_dict = {}
    for line in txt.splitlines():
        if line.startswith('#') or not line.strip():
            continue
        parts = re.split(r'\s+', line.strip())
        if len(parts) >= 4 and parts[0].startswith('rs'):
            rsid = parts[0]
            genotype = parts[3].upper().replace(' ', '')
            snp_dict[rsid] = genotype if len(genotype) <= 2 else '--'
    return snp_dict


snp_map = parse_23andme(raw_text)
print(f"‚úì Parsed {len(snp_map):,} SNPs from your genome\n")

# ‚îÄ‚îÄ 7. PGS Calculation ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ


def count_effect_alleles(genotype, effect_allele):
    """Count effect alleles (0, 1, or 2)"""
    if not genotype or genotype in {'--', 'NN', '0', '', 'DI', 'ID'}:
        return None
    gt = genotype.strip().upper()
    effect = effect_allele.upper()
    if len(gt) == 1:
        return 2 if gt == effect else 0
    return sum(c == effect for c in gt)


def compute_pgs(snp_map, snp_db, selected_ancestry='EUR'):
    """Calculate evidence-based PGS with ancestry-specific effects"""
    print(
        f"üîç DEBUG: compute_pgs called with selected_ancestry = '{selected_ancestry}'")

    contributions = {}
    raw_score = 0.0
    n_valid = 0
    total_weight = 0.0
    ancestry_adjustments = []

    for rsid, meta in snp_db.items():
        genotype = snp_map.get(rsid, None)
        dosage = count_effect_alleles(genotype, meta['effect_allele'])

        # Get ancestry-specific beta
        if meta.get('ancestry_specific', False):
            # This SNP has ancestry-specific effects
            beta_key = f'beta_{selected_ancestry}'
            if beta_key in meta:
                beta = meta[beta_key]
                ancestry_adjustments.append(
                    f"{rsid}: Using {selected_ancestry}-specific beta = {beta:+.4f}")
                # Special debug for rs4680
                if rsid == 'rs4680':
                    print(
                        f"üîç DEBUG rs4680: ancestry_specific=True, beta_key='{beta_key}', beta={beta}")
            else:
                beta = meta.get('beta_default', 0.025)
                ancestry_adjustments.append(
                    f"{rsid}: No {selected_ancestry}-specific data, using default beta = {beta:+.4f}")
        else:
            # Use ancestry-specific or default
            beta_key = f'beta_{selected_ancestry}'
            beta = meta.get(beta_key, meta.get('beta_default', 0.025))
            if rsid == 'rs4680':
                print(
                    f"üîç DEBUG rs4680: ancestry_specific=False, beta_key='{beta_key}', beta={beta}")

        contrib = dosage * beta if dosage is not None else None

        contributions[rsid] = {
            'genotype': genotype or '‚Äî',
            'dosage': dosage,
            'beta': beta,
            'contrib': contrib,
            'gene': meta['gene'],
            'evidence': meta['evidence'],
            'phenotype': meta['phenotype']
        }

        if contrib is not None:
            raw_score += contrib
            n_valid += 1
            total_weight += abs(beta)

    # Print ancestry adjustments for key SNPs
    if ancestry_adjustments:
        print(f"\nüß¨ Ancestry-specific adjustments for {selected_ancestry}:")
        for adj in ancestry_adjustments:
            print(f"   {adj}")
        print()

    return raw_score, contributions, n_valid, total_weight


raw_score, contributions, n_valid_snps, total_weight = compute_pgs(
    snp_map, SNP_DB_PGS, SELECTED_ANCESTRY)

# Show example of ancestry effect (if rs4680 is present)
if 'rs4680' in snp_map and snp_map['rs4680'] not in {'--', 'NN', '0', '', 'DI', 'ID'}:
    rs4680_genotype = snp_map['rs4680']
    rs4680_dosage = count_effect_alleles(rs4680_genotype, 'A')
    if rs4680_dosage is not None:
        eur_contrib = rs4680_dosage * 0.042
        eas_contrib = rs4680_dosage * (-0.042)
        diff = abs(eur_contrib - eas_contrib)
        print(
            f"üß¨ ANCESTRY EFFECT EXAMPLE (rs4680): Genotype {rs4680_genotype}")
        print(f"   EUR ancestry contribution: {eur_contrib:+.4f}")
        print(f"   EAS ancestry contribution: {eas_contrib:+.4f}")
        print(f"   Absolute difference: {diff:.4f} points")
        print(
            f"   ‚úì Using {SELECTED_ANCESTRY} = {contributions['rs4680']['contrib']:+.4f}")
        print()

# ‚îÄ‚îÄ 8. Standardize to Z-score ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Population parameters based on allele frequencies
# Assuming average MAF ~0.25 for these selected SNPs


def calculate_population_parameters(snp_db, selected_ancestry='EUR', maf=0.25):
    """Calculate expected population mean and SD using ancestry-specific betas"""
    expected_mean = 0.0
    expected_var = 0.0

    for meta in snp_db.values():
        # Get ancestry-specific beta
        if meta.get('ancestry_specific', False):
            beta_key = f'beta_{selected_ancestry}'
            beta = meta.get(beta_key, meta.get('beta_default', 0.025))
        else:
            beta_key = f'beta_{selected_ancestry}'
            beta = meta.get(beta_key, meta.get('beta_default', 0.025))

        expected_mean += 2 * maf * beta
        expected_var += 2 * maf * (1 - maf) * beta**2

    expected_sd = np.sqrt(expected_var)
    return expected_mean, expected_sd


pop_mean, pop_sd = calculate_population_parameters(
    SNP_DB_PGS, SELECTED_ANCESTRY)

# Calculate z-score and percentile
z_score = (raw_score - pop_mean) / pop_sd if pop_sd > 0 else 0
percentile = 100 * norm.cdf(z_score)

# Estimate variance explained (R¬≤)
# Conservative estimate based on number of genome-wide significant SNPs
n_gwas_sig = sum(1 for meta in SNP_DB_PGS.values()
                 if meta.get('p_value', 1) < 5e-8)
estimated_r2 = min(0.001 * n_valid_snps + 0.005 *
                   n_gwas_sig, 0.05)  # Cap at 5%

print(f"‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó")
print(f"‚ïë           POLYGENIC SCORE RESULTS                    ‚ïë")
print(f"‚ï†‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ï£")
print(f"‚ïë  Raw Score:        {raw_score:8.4f}                      ‚ïë")
print(f"‚ïë  Z-Score:          {z_score:+8.3f}                      ‚ïë")
print(f"‚ïë  Percentile:       {percentile:7.1f}%                      ‚ïë")
print(
    f"‚ïë  Valid SNPs:       {n_valid_snps:2d} / {len(SNP_DB_PGS):2d}                          ‚ïë")
print(f"‚ïë  Est. R¬≤ (max):    {estimated_r2*100:7.2f}%                      ‚ïë")
print(f"‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù\n")

# ‚îÄ‚îÄ 9. Detailed SNP Table ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
rows = []
for rsid, meta in SNP_DB_PGS.items():
    c = contributions[rsid]
    ancestry_flag = " *" if meta.get('ancestry_specific', False) else ""
    rows.append({
        'SNP': rsid + ancestry_flag,
        'Gene': meta['gene'],
        'Genotype': c['genotype'],
        'Dosage': c['dosage'] if c['dosage'] is not None else '‚Äî',
        'Beta': f"{c['beta']:+.4f}",
        'Contribution': f"{c['contrib']:+.5f}" if c['contrib'] is not None else '‚Äî',
        'Evidence': meta['evidence'],
        'Phenotype': meta['phenotype'][:40] + '...' if len(meta['phenotype']) > 40 else meta['phenotype']
    })

df_table = pd.DataFrame(rows)
display(HTML(
    f"<h3>üìä Individual SNP Contributions (Using {SELECTED_ANCESTRY} ancestry betas)</h3>"))
display(df_table.style.set_properties(**{
    'text-align': 'left',
    'font-size': '11pt'
}).set_table_styles([
    {'selector': 'th', 'props': [
        ('font-size', '11pt'), ('font-weight', 'bold')]}
]))

# Add legend for ancestry-specific markers
n_ancestry_specific = sum(1 for meta in SNP_DB_PGS.values()
                          if meta.get('ancestry_specific', False))
if n_ancestry_specific > 0:
    display(HTML(f"""
    <p style="font-size: 10pt; color: #666; margin-top: 5px;">
    <b>* = Ancestry-specific SNP</b> ({n_ancestry_specific} SNPs have different effects across ancestries)
    </p>
    """))

# ‚îÄ‚îÄ 10. Visualizations ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. Population Distribution
ax1 = fig.add_subplot(gs[0, :2])
x = np.linspace(-3, 3, 500)
pop_pdf = norm.pdf(x, 0, 1)
ax1.plot(x, pop_pdf, color="#2c3e50", lw=3, label="Population Distribution")
ax1.axvline(z_score, color="#e74c3c", lw=2.5, ls="--",
            label=f"Your Score (z={z_score:+.2f})")
ax1.fill_between(x, pop_pdf, where=(x <= z_score), alpha=0.25, color="#e74c3c")
ax1.set(title="Your Genetic Score vs Population",
        xlabel="Z-score", ylabel="Probability Density")
ax1.legend(fontsize=10)
ax1.grid(alpha=0.3)

# 2. Percentile Bar
ax2 = fig.add_subplot(gs[0, 2])
color = "#27ae60" if percentile > 60 else "#e74c3c" if percentile < 40 else "#3498db"
ax2.barh(["Your\nScore"], [percentile], color=color,
         height=0.5, edgecolor='black', linewidth=1.5)
ax2.set_xlim(0, 100)
ax2.axvline(50, color='gray', ls='--', alpha=0.5, lw=2)
ax2.set_title(f"Percentile Rank", fontweight='bold')
ax2.set_xlabel("Percentile")
for i in [25, 50, 75]:
    ax2.axvline(i, color='gray', ls=':', alpha=0.3)
ax2.text(percentile + 3, 0, f"{percentile:.1f}%",
         va='center', fontsize=13, fontweight='bold')

# 3. SNP Contributions
ax3 = fig.add_subplot(gs[1, :])
contrib_data = []
for rsid, meta in SNP_DB_PGS.items():
    c = contributions[rsid]
    if c['contrib'] is not None:
        contrib_data.append({
            'SNP': f"{rsid}\n({meta['gene']})",
            'Contribution': c['contrib'],
            'Evidence': meta['evidence']
        })

df_bar = pd.DataFrame(contrib_data).sort_values('Contribution')
colors_map = {'Strong': '#27ae60', 'Moderate': '#f39c12', 'Weak': '#95a5a6'}
bar_colors = [colors_map.get(e, '#95a5a6') for e in df_bar['Evidence']]

bars = ax3.barh(df_bar['SNP'], df_bar['Contribution'],
                color=bar_colors, edgecolor='black', linewidth=0.8)
ax3.axvline(0, color='black', lw=1)
ax3.set_title("Individual SNP Contributions to Your PGS",
              fontweight='bold', fontsize=12)
ax3.set_xlabel("Weighted Contribution (Beta √ó Dosage)")
ax3.grid(axis='x', alpha=0.3)

# Legend for evidence strength
legend_elements = [Patch(facecolor='#27ae60', label='Strong Evidence'),
                   Patch(facecolor='#f39c12', label='Moderate Evidence'),
                   Patch(facecolor='#95a5a6', label='Weak Evidence')]
ax3.legend(handles=legend_elements, loc='lower right', fontsize=9)

# 4. Evidence Strength Pie
ax4 = fig.add_subplot(gs[2, 0])
evidence_counts = df_table['Evidence'].value_counts()
colors_pie = [colors_map.get(e, '#95a5a6') for e in evidence_counts.index]
ax4.pie(evidence_counts.values, labels=evidence_counts.index, autopct='%1.0f%%',
        colors=colors_pie, startangle=90, textprops={'fontsize': 10, 'fontweight': 'bold'})
ax4.set_title("Evidence Quality Distribution", fontweight='bold')

# 5. Phenotype Categories
ax5 = fig.add_subplot(gs[2, 1:])
phenotype_categories = {
    'Reading/Language': ['Nonword reading', 'Phonemic awareness', 'reading'],
    'Memory': ['memory', 'Hippocampal'],
    'Auditory': ['Hearing', 'auditory'],
    'Cognition': ['IQ', 'cognitive', 'attainment', 'working memory'],
    'Dopamine': ['COMT', 'DRD2', 'creative']
}

category_scores = {}
for cat, keywords in phenotype_categories.items():
    cat_score = 0
    for rsid, meta in SNP_DB_PGS.items():
        if any(kw.lower() in meta['phenotype'].lower() for kw in keywords):
            contrib = contributions[rsid]['contrib']
            if contrib is not None:
                cat_score += contrib
    category_scores[cat] = cat_score

cats = list(category_scores.keys())
scores = list(category_scores.values())
colors_cat = ['#3498db', '#9b59b6', '#e67e22', '#1abc9c', '#e74c3c']
bars_cat = ax5.barh(cats, scores, color=colors_cat,
                    edgecolor='black', linewidth=1)
ax5.axvline(0, color='black', lw=1)
ax5.set_title("Contributions by Cognitive Domain", fontweight='bold')
ax5.set_xlabel("Cumulative Contribution")
ax5.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# ‚îÄ‚îÄ 11. Interpretation ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ


def interpret_score(z, pct, r2, n_valid):
    """Generate evidence-based interpretation"""

    # Determine category
    if z < -1.5:
        cat = "Well Below Average"
        color = "#c0392b"
    elif z < -0.5:
        cat = "Below Average"
        color = "#e67e22"
    elif z < 0.5:
        cat = "Average"
        color = "#3498db"
    elif z < 1.5:
        cat = "Above Average"
        color = "#27ae60"
    else:
        cat = "Well Above Average"
        color = "#16a085"

    # Generate interpretation
    if z < -1:
        main_text = f"""Your polygenic score is in the <b>bottom {100-pct:.0f}%</b> of the population for these
        language-related genetic variants. However, this is a <b>tiny factor</b> in language learning success."""

        advice = """
        <b>What this means for you:</b>
        <ul>
        <li>You may need slightly more repetitions to form strong phonological memories</li>
        <li>Focus on <b>systematic phonics training</b> and pronunciation drills</li>
        <li>Use <b>spaced repetition aggressively</b> (Anki/SRS daily)</li>
        <li>Consider spending 10-15% more time on explicit grammar study</li>
        <li><b>Critical point:</b> Someone at the 10th percentile who studies 2 hours/day will far surpass
        someone at the 90th percentile who studies 30 min/day</li>
        </ul>
        """

    elif z > 1:
        main_text = f"""Your polygenic score is in the <b>top {100-pct:.0f}%</b> for these variants.
        This may provide a <b>slight efficiency advantage</b>, but it's still a minor factor."""

        advice = """
        <b>What this means for you:</b>
        <ul>
        <li>You may acquire phonological patterns slightly faster</li>
        <li>You can likely handle more <b>immersion-heavy approaches</b> (70%+ input)</li>
        <li>Don't skip the fundamentals - genetics ‚â† automatic fluency</li>
        <li>Your advantage is <b>learning rate</b>, not ultimate attainment</li>
        <li><b>Key insight:</b> This 2-4% genetic edge can be easily lost to poor study methods or low motivation</li>
        </ul>
        """
    else:
        main_text = f"""Your polygenic score is <b>average</b> (around the 50th percentile).
        This is where most successful language learners are."""

        advice = """
        <b>What this means for you:</b>
        <ul>
        <li>Use evidence-based methods: comprehensible input + spaced repetition + production practice</li>
        <li>Balance immersion (60%) with explicit study (25%) and speaking practice (15%)</li>
        <li>Consistency matters far more than genetics</li>
        <li>Track your hours - aim for 600-1200 hours for B2-level proficiency</li>
        </ul>
        """

    # R¬≤ interpretation
    variance_text = f"""
    <div style="background-color: #ecf0f1; padding: 10px; border-radius: 5px; margin-top: 10px;">
    <b>üìä Effect Size Context:</b><br>
    These {n_valid} genetic variants explain an estimated <b>~{r2*100:.1f}% of variance</b> in related cognitive traits.
    This means <b>{100-r2*100:.1f}%</b> comes from:<br>
    ‚Ä¢ Study method & quality (30-40%)<br>
    ‚Ä¢ Total practice hours (20-30%)<br>
    ‚Ä¢ Motivation & consistency (15-20%)<br>
    ‚Ä¢ Age of acquisition (5-10%)<br>
    ‚Ä¢ Other environmental factors (10-15%)<br>
    ‚Ä¢ Measurement error & unmeasured genetics (~{r2*100:.1f}%)
    </div>
    """

    return cat, color, main_text, advice, variance_text


category, cat_color, main_interp, advice_text, variance_interp = interpret_score(
    z_score, percentile, estimated_r2, n_valid_snps)

display(HTML(f"""
<div style="background-color: {cat_color}15; border-left: 5px solid {cat_color}; padding: 20px; margin: 20px 0;">
<h2 style="color: {cat_color}; margin-top: 0;">Your Result: {category}</h2>
<p style="font-size: 14pt;">{main_interp}</p>
{advice_text}
{variance_interp}
</div>
"""))

# ‚îÄ‚îÄ 12. Gene-by-Gene Insights ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(HTML("<h3>üß¨ Gene-by-Gene Insights (Top Contributors)</h3>"))

top_contributors = sorted(
    [(rsid, contributions[rsid], SNP_DB_PGS[rsid])
     for rsid in SNP_DB_PGS.keys()
     if contributions[rsid]['contrib'] is not None],
    key=lambda x: abs(x[1]['contrib']),
    reverse=True
)[:5]

for rsid, contrib, meta in top_contributors:
    refs_list = ", ".join([f"<a href='{REFERENCES[ref].get('url', '#')}' target='_blank'>{ref}</a>"
                           for ref in meta['refs'] if ref in REFERENCES])

    dosage = contrib['dosage']
    if dosage == 2:
        dosage_text = "üî¥üî¥ Homozygous for effect allele (2 copies)"
    elif dosage == 1:
        dosage_text = "üü° Heterozygous (1 copy)"
    else:
        dosage_text = "‚ö™ No effect allele copies"

    display(HTML(f"""
    <div style="background-color: #f8f9fa; border-left: 3px solid #3498db; padding: 12px; margin: 10px 0; border-radius: 3px;">
    <b style="font-size: 12pt;">{rsid}</b> ({meta['gene']}) - {dosage_text}<br>
    <b>Your genotype:</b> {contrib['genotype']} | <b>Contribution:</b> {contrib['contrib']:+.5f}<br>
    <b>Phenotype:</b> {meta['phenotype']}<br>
    <b>Evidence:</b> {meta['evidence']} | <b>Study:</b> {meta['population']}<br>
    <b>References:</b> {refs_list}<br>
    <i style="font-size: 10pt; color: #7f8c8d;">{meta['notes']}</i>
    </div>
    """))

# ‚îÄ‚îÄ 13. Ancestry Warning ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(HTML("""
<div style="background-color: #fff3cd; border-left: 5px solid #ff9800; padding: 15px; margin: 20px 0;">
<h3 style="color: #e65100; margin-top: 0;">‚ö†Ô∏è CRITICAL: Ancestry Considerations</h3>
<p style="color: #e65100; font-weight: bold;">These effect sizes are primarily from European-ancestry samples.
Predictions may be INACCURATE for other ancestries:</p>
<table style="width: 100%; color: #e65100;">
<tr><td><b>East Asian ancestry:</b></td><td>COMT rs4680 shows <b>OPPOSITE effects</b> (Val > Met for working memory)</td></tr>
<tr><td><b>African ancestry:</b></td><td>Different linkage disequilibrium patterns make SNP-level predictions unreliable</td></tr>
<tr><td><b>Mixed ancestry:</b></td><td>Effects are unpredictable without ancestry-specific analysis</td></tr>
<tr><td><b>Non-European:</b></td><td>Most loci have not been validated; transferability is poor</td></tr>
</table>
<p style="color: #e65100; margin-top: 10px;"><b>Bottom line:</b> If you're not of primarily European descent,
treat these results as <i>highly speculative</i>.</p>
</div>
"""))

# ‚îÄ‚îÄ 14. Evidence-Based Learning Plan ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(Markdown(f"""
## üìö Evidence-Based Language Learning Plan
### (Based on research, not your genetics)

Your genetic score is **{category.lower()}**, but this changes almost nothing about the optimal study approach.
Here's what actually works:

### üéØ The 70-20-10 Framework (2 hours/day)

**1. Comprehensible Input (70% = 84 min/day)**
- **Listening:** Podcasts, YouTube, audiobooks at 90% comprehension level
- **Reading:** Graded readers ‚Üí native materials with popup dictionaries
- **Focus:** Volume over perfection. Aim for 10+ hours/week of input
- **Why it works:** Input builds implicit grammar and vocabulary automatically

**2. Explicit Study (20% = 24 min/day)**
- **SRS (Anki):** 15 min/day of sentence mining or frequency-based decks
- **Grammar:** 10 min/day learning patterns in context (not rules in isolation)
- **Pronunciation:** 5 min shadowing or phonetic drills
- **Why it works:** Accelerates pattern recognition that input alone misses

**3. Production Practice (10% = 12 min/day)**
- **Speaking:** iTalki tutors, language exchange, or shadowing
- **Writing:** Journaling with corrections (free services: r/WriteStreak, HelloTalk)
- **Why it works:** Forces retrieval, exposes gaps, builds fluency

---

### üìÖ Weekly Schedule Example

| Day | Morning (30 min) | Evening (90 min) |
|-----|------------------|------------------|
| **Mon** | Anki review (15 min) + Grammar (15 min) | Podcast (60 min) + Shadowing (30 min) |
| **Tue** | Anki (15 min) + Reading (15 min) | YouTube immersion (90 min) |
| **Wed** | Anki (15 min) + Pronunciation (15 min) | Tutor session (60 min) + Review (30 min) |
| **Thu** | Anki (15 min) + Grammar (15 min) | Reading (60 min) + Listening (30 min) |
| **Fri** | Anki (15 min) + Writing (15 min) | Movie/TV series (90 min) |
| **Sat** | Rest or light review | Long immersion (3-4 hours) |
| **Sun** | Anki (20 min) | Conversation practice (60 min) + Planning (10 min) |

**Total:** ~14 hours/week = B2 level in 10-18 months

---

### üî¨ What the Research Actually Shows

**Factors that predict success (sorted by effect size):**
1. ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê **Total hours of practice** (R¬≤ ‚âà 0.40-0.60)
2. ‚≠ê‚≠ê‚≠ê‚≠ê **Quality of input** (comprehensibility) (R¬≤ ‚âà 0.15-0.25)
3. ‚≠ê‚≠ê‚≠ê **Age of acquisition** (R¬≤ ‚âà 0.10-0.20 for pronunciation)1
4. ‚≠ê‚≠ê **Working memory** (R¬≤ ‚âà 0.05-0.10)
5. ‚≠ê‚≠ê **Motivation & persistence** (R¬≤ ‚âà 0.05-0.10)
6. ‚≠ê **Your genetic variants** (R¬≤ ‚âà {estimated_r2:.3f})

**Translation:** If you score at the **10th genetic percentile** but study 2 hours/day with good methods,
you'll surpass someone at the **90th percentile** who studies 30 min/day poorly.

---

### üéì Genetic Modifiers (Minor Adjustments)

{"**If your score is below average (<40th percentile):**" if percentile < 40 else "**If your score is above average (>60th percentile):**" if percentile > 60 else "**Your score is average:**"}

{'''
- Add +10 min/day to explicit phonics/pronunciation work
- Use more structured courses initially (e.g., Pimsleur, Assimil)
- Front-load grammar study in first 3 months
- Accept that you'll need 10-20% more repetitions''' if percentile < 40 else '''
- Can handle more immersion-heavy approaches (75% input)
- May grasp patterns faster, but don't skip fundamentals
- Use your efficiency to learn multiple languages or go deeper
- Still need 600+ hours for fluency - no shortcuts''' if percentile > 60 else '''
- Use the standard 70-20-10 framework
- Adjust based on personal preference and results
- Track hours and progress metrics
- Focus on consistency over optimization'''}

**Key insight:** Genetic differences might mean you need 800 hours vs. 700 hours to reach B2.
Study method determines whether you need 700 vs. 2000 hours.

---

### üìä Success Metrics (Track These)

- **Primary:** Total hours logged (aim for 600-1200 for B2)
- **Secondary:** Anki cards reviewed, pages read, podcast episodes completed
- **Outcome:** Can you understand 80%+ of native content? Pass a proficiency test?

**Genetics tell you almost nothing. Your tracking data tells you everything.**
"""))

# ‚îÄ‚îÄ 15. Interactive Scenarios ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(HTML("<h3>üéÆ Interactive: Compare Learning Scenarios</h3>"))


def scenario_calculator(daily_minutes, method_quality, consistency, genetic_percentile):
    """
    Estimate time to B2 based on actual important factors
    genetic_percentile has minimal effect (realistic)
    """
    # Base hours needed for B2
    base_hours = 800

    # Method quality multiplier (largest effect)
    if method_quality == "Optimal":
        method_mult = 0.85  # 15% faster
    elif method_quality == "Good":
        method_mult = 1.0
    elif method_quality == "Poor":
        method_mult = 1.5  # 50% slower
    else:
        method_mult = 2.0  # "Terrible" - grammar-translation only

    # Consistency multiplier
    if consistency == "High":
        consist_mult = 0.95
    elif consistency == "Medium":
        consist_mult = 1.05
    else:
        consist_mult = 1.25

    # Genetic multiplier (MINIMAL effect - realistic)
    if genetic_percentile >= 80:
        genetic_mult = 0.97  # 3% faster
    elif genetic_percentile >= 60:
        genetic_mult = 0.99  # 1% faster
    elif genetic_percentile <= 20:
        genetic_mult = 1.05  # 5% slower
    elif genetic_percentile <= 40:
        genetic_mult = 1.02  # 2% slower
    else:
        genetic_mult = 1.0

    # Calculate
    adjusted_hours = base_hours * method_mult * consist_mult * genetic_mult
    daily_hours = daily_minutes / 60
    months = (adjusted_hours / daily_hours) / 30

    return adjusted_hours, months


scenarios = (
    ("You (current genetics)", percentile, "Good", 120, "High"),
    ("Same genetics, poor method", percentile, "Poor", 120, "High"),
    ("Same genetics, low consistency", percentile, "Good", 120, "Low"),
    ("Top 10% genetics, poor method", 95, "Poor", 120, "High"),
    ("Bottom 10% genetics, optimal method", 5, "Optimal", 120, "High"),
    ("Average genetics, 30 min/day", 50, "Good", 30, "Medium")
)

results = []
for name, genetic_pct, method, mins, consist in scenarios:
    hours, months = scenario_calculator(mins, method, consist, genetic_pct)
    results.append({
        'Scenario': name,
        'Genetics': f"{genetic_pct:.0f}th %ile",
        'Method': method,
        'Daily Min': mins,
        'Consistency': consist,
        'Total Hours': f"{hours:.0f}",
        'Months to B2': f"{months:.1f}"
    })

df_scenarios = pd.DataFrame(results)
display(HTML("<p><b>Time to reach B2 level (conversational fluency) under different scenarios:</b></p>"))
display(df_scenarios.style.set_properties(**{
    'text-align': 'center',
    'font-size': '11pt'
}).set_table_styles([
    {'selector': 'th', 'props': [
        ('font-size', '11pt'), ('font-weight', 'bold'), ('text-align', 'center')]}
]).background_gradient(subset=['Months to B2'], cmap='RdYlGn_r'))

display(HTML("""
<p style="margin-top: 15px; padding: 10px; background-color: #e8f5e9; border-left: 4px solid #4caf50;">
<b>üîç Key Insight:</b> Notice how \"Bottom 10% genetics + optimal method\" beats \"Top 10% genetics + poor method\" by <b>10+ months</b>.
This is the real-world importance of genetics vs. behavior for language learning.
</p>
"""))

# ‚îÄ‚îÄ 16. Full Reference List with Links ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(HTML("<h3>üìñ Complete References</h3>"))

ref_html = "<div style='max-height: 400px; overflow-y: scroll; border: 1px solid #ddd; padding: 15px; background-color: #f9f9f9;'>"
for ref_key, ref_data in sorted(REFERENCES.items()):
    ref_html += f"""
    <p style='margin-bottom: 15px;'>
    <b>{ref_key}:</b> {ref_data['citation']}<br>
    <a href=\"{ref_data.get('url', '#')}\" target=\"_blank\" style=\"color: #2980b9;\">{ref_data.get('url', 'URL N/A')}</a>
    {f"<br><span style='font-size: 10pt; color: #7f8c8d;'>PMID: {ref_data['pmid']}</span>" if 'pmid' in ref_data else ""}
    <br><i style=\"font-size: 10pt; color: #27ae60;\">{ref_data['key_finding']}</i>
    </p>
    """
ref_html += "</div>"
display(HTML(ref_html))

# ‚îÄ‚îÄ 17. JSON Export ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
report = {
    "metadata": {
        "generated": datetime.now(timezone.utc).isoformat(),
        "version": "3.5_evidence_based",
        "disclaimer": "Educational tool only. Not diagnostic or predictive of individual language learning success."
    },
    "pgs_results": {
        "raw_score": round(raw_score, 5),
        "z_score": round(z_score, 3),
        "percentile": round(percentile, 1),
        "category": category,
        "estimated_r2_percent": round(estimated_r2 * 100, 2),
        "n_valid_snps": n_valid_snps,
        "n_total_snps": len(SNP_DB_PGS)
    },
    "snp_contributions": {
        rsid: {
            'gene': SNP_DB_PGS[rsid]['gene'],
            'genotype': contributions[rsid]['genotype'],
            'dosage': contributions[rsid]['dosage'],
            'beta': contributions[rsid]['beta'],
            'contribution': contributions[rsid]['contrib'],
            'evidence': contributions[rsid]['evidence'],
            'phenotype': contributions[rsid]['phenotype']
        } for rsid in SNP_DB_PGS.keys()
    },
    "interpretation": {
        "category": category,
        "main_text": main_interp,
        "advice": advice_text
    },
    "scenarios": df_scenarios.to_dict(orient='records'),
    "references": REFERENCES,
    "warnings": [
        "No validated PGS for language learning exists",
        f"These variants explain only ~{estimated_r2*100:.1f}% of variance in related traits",
        "Environmental factors (study method, hours, motivation) are 20-50x more important",
        "Effect sizes are ancestry-specific; most data is from European populations",
        "Cross-trait inference: using reading/memory/hearing genetics as proxy",
        "For educational/research purposes only - not clinical or diagnostic"
    ]
}

fname = f"genelingua_report_v3.5_{datetime.now().strftime('%Y%m%d')}.json"
with open(fname, "w") as f:
    json.dump(report, f, indent=2)
files.download(fname)

display(HTML(f"""
<div style="margin-top: 30px; padding: 20px; background-color: #e3f2fd; border-radius: 5px; border-left: 5px solid #2196F3;">
<h3 style="margin-top: 0; color: #1976D2;">‚úÖ Analysis Complete</h3>
<p><b>Report downloaded:</b> <code>{fname}</code></p>
<p style="margin-top: 15px; font-size: 13pt; line-height: 1.6;">
<b>Remember:</b> This tool demonstrates polygenic scoring methodology using published genetic associations
with reading, memory, hearing, and cognitive ability. While these traits correlate with language learning,
<b>no validated genetic predictor of language learning success exists</b>.<br><br>

Your success will be determined by:<br>
‚Ä¢ <b>How you study</b> (70-20-10 method) ‚Üí 30-40% of variance<br>
‚Ä¢ <b>How many hours</b> you log (600-1200 for B2) ‚Üí 20-30% of variance<br>
‚Ä¢ <b>Your motivation</b> and consistency ‚Üí 15-20% of variance<br>
‚Ä¢ <b>These genetic variants</b> ‚Üí ~{estimated_r2*100:.1f}% of variance<br><br>

<b style="color: #1976D2; font-size: 14pt;">Go study. Your genetics are fine. üöÄ</b>
</p>
</div>
"""))

# ‚îÄ‚îÄ 18. Ancestry Logic Test ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ


def test_ancestry_logic():
    """Test function to verify ancestry-specific beta handling"""
    print("\n" + "="*60)
    print("üß™ TESTING ANCESTRY-SPECIFIC LOGIC")
    print("="*60)

    # Test with a fake rs4680 genotype
    test_snp_map = {'rs4680': 'AG'}  # 1 copy of A (Met) allele

    print("Testing rs4680 (AG genotype = 1 Met allele):")
    print("Expected results:")
    print("  EUR ancestry: 1 √ó (+0.042) = +0.042")
    print("  EAS ancestry: 1 √ó (-0.042) = -0.042")
    print("  Difference: 0.084 points")
    print()

    # Test EUR
    print("üß¨ Testing EUR ancestry:")
    score_eur, contrib_eur, _, _ = compute_pgs(
        test_snp_map, {'rs4680': SNP_DB_PGS['rs4680']}, 'EUR')
    rs4680_contrib_eur = contrib_eur['rs4680']['contrib']
    print(f"   Result: {rs4680_contrib_eur:+.4f}")

    # Test EAS
    print("\nüß¨ Testing EAS ancestry:")
    score_eas, contrib_eas, _, _ = compute_pgs(
        test_snp_map, {'rs4680': SNP_DB_PGS['rs4680']}, 'EAS')
    rs4680_contrib_eas = contrib_eas['rs4680']['contrib']
    print(f"   Result: {rs4680_contrib_eas:+.4f}")

    # Check difference
    diff = abs(rs4680_contrib_eur - rs4680_contrib_eas)
    print(f"\nüìä Difference: {diff:.4f}")

    if diff > 0.08:  # Should be 0.084
        print("‚úÖ SUCCESS: Ancestry-specific logic is working!")
        print("   EUR and EAS show different effects for rs4680")
    else:
        print("‚ùå FAILED: Ancestry logic is not working")
        print("   EUR and EAS show same effects - something is wrong")

    print("="*60)


# Run the test
test_ancestry_logic()

print("\n" + "="*60)
print("  ANALYSIS COMPLETE - Now go practice your target language!")
print("="*60)
```
